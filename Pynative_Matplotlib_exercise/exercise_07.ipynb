### Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
total_profit = df['total_profit'].to_list()
xlist = np.arange(150000, 350001, 25000).tolist()
```

```python
plt.hist(total_profit, xlist, label='Profit data')
plt.legend(loc='upper left')
plt.title('Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.xticks(xlist)
plt.show()
```

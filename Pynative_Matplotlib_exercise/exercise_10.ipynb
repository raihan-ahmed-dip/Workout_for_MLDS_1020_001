### Read all product sales data and show it using the stack plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
months = df['month_number'].to_list()
face_cream = df['facecream'].to_list()
face_wash = df['facewash'].to_list()
toothpaste = df['toothpaste'].to_list()
bathing_soap = df['bathingsoap'].to_list()
shampoo = df['shampoo'].to_list()
moisturizer = df['moisturizer'].to_list()
```

```python
plt.stackplot(months, face_cream, face_wash, toothpaste, bathing_soap, shampoo, moisturizer, labels=['face cream', 'Face wash', 'Tooth paste', 'Bathing soap', 'Shampoo', 'Moisturizer'])
plt.title('All product sales data using stack plot')
plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')
plt.legend(loc='upper left')
plt.xticks(months)
plt.yticks(np.arange(0, 30001, 5000).tolist())
plt.show()
```

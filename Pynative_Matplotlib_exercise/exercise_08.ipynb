### Calculate total sale data for last year for each product and show it using a Pie chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
face_cream = df['facecream'].sum()
face_wash = df['facewash'].sum()
toothpaste = df['toothpaste'].sum()
bathing_soap = df['bathingsoap'].sum()
shampoo = df['shampoo'].sum()
moisturizer = df['moisturizer'].sum()
```

```python
productwise_sales_data = {
    'product': [face_cream, face_wash, toothpaste, bathing_soap, shampoo, moisturizer],
    'name': ['FaceCream', 'FaceWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
}
```

```python
plt.pie(productwise_sales_data['product'], labels=productwise_sales_data['name'], autopct='%1.1f%%')
plt.title('Sales data')
plt.legend(loc='lower right')
plt.axis("equal")
plt.show()
```

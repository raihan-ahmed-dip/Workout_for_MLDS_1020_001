### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
months = df['month_number'].to_list()
bathing_soap = df['bathingsoap'].to_list()
face_wash = df['facewash'].to_list()
```

```python
n, sub = plt.subplots(2, sharex=True)
sub[0].plot(months, bathing_soap, color='k', marker='o')
sub[0].set_title('Sales data of a Bathingsoap')
sub[1].plot(months, face_wash, color='r', marker='o')
sub[1].set_title('Sales data of a facewash')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(months)
plt.show()
```

### Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
months = df['month_number'].to_list()
bathing_soap = df['bathingsoap'].to_list()

ylist = np.arange(0, 14001, 2000).tolist()
```

```python
plt.bar(months, bathing_soap)
plt.title('bathingsoap sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(months)
plt.yticks(ylist)
plt.grid(True, linestyle='--')
plt.savefig('bathingsoap_sales_data.png', dpi=600)
plt.show()
```

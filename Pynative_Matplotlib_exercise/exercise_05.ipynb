### Read face cream and facewash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
months = df['month_number'].to_list()
face_cream = df['facecream'].to_list()
face_wash = df['facewash'].to_list()

ylist = np.arange(0, 3501, 500).tolist()
```

```python
plt.bar([m-0.25 for m in months], face_cream, label='Face Cream sales data', align='edge', width=0.25)
plt.bar(months, face_wash, label='Face Wash sales data', align='edge', width=0.25)
plt.title('Facewash and Facecream sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(months)
plt.yticks(ylist)
plt.grid(True, linestyle='--')
plt.show()
```
